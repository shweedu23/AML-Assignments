```python
df = pd.read_csv('weight.csv')
```

```python

x = df.iloc[:,2]
y = df.iloc[:,1]
n = len(y)
```

```python
x1 = (x- x.mean())/ x.std()
x1
```

```python
X = np.c_[np.ones(n),x1]
X
```

```python
plt.plot(x1,y,'.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
theta
```

```python
plt.plot(x,y,'o')
plt.plot(x, theta[0] + theta[1]*x1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
